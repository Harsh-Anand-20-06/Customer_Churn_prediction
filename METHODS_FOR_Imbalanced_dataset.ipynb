#### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
y = df["Churn"]
x = df.drop("Churn",axis = 1)
```

```python
y.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15,stratify=y)
## never touch test sample, apply smote on train sample only
```

```python
smote = SMOTE(sampling_strategy="minority")
x_sm_train,y_sm_train = smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
y_sm_train.value_counts()
```

```python
model = Sequential([
    keras.layers.Dense(10,input_shape=(26,),activation="relu"),
    keras.layers.Dense(5,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")   ## We needed 10 neurons in last layer in digit dataset,here 1 because, there we needed all
])                                               ## 10 neurons output to get max of all and predict class, here, whatever this 1 neuron gives
                                                ## we can predict on basis of that only, we do not need 2 neurons explicitly
model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)

model.fit(x_sm_train,y_sm_train,epochs = 50)
```

```python
y_pred = model.predict(x_test)

for idx in range(len(y_pred)):
    if y_pred[idx]>=0.5:
        y_pred[idx]=1
    else:
        y_pred[idx]=0


```

```python
print(classification_report(y_test,y_pred))
```
